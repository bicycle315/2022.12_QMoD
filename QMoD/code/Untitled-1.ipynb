```python
from qutip import coherent, basis, fock, plot_wigner, plot_wigner_fock_distribution
import numpy as np
from matplotlib import pyplot as plt
from qutip import *
```

```python
catplus=(coherent(20,6)+coherent(20,-6))
plot_wigner_fock_distribution(catplus)
catminus=(coherent(20,6)-coherent(20,-6))
plot_wigner_fock_distribution(catminus)
```

```python
ani_CP=ani*catplus
wigner_fock_distribution(ani_CP)
ani_CM=ani*catminus
wigner_fock_distribution(ani_CM)
```

```python
cre=create(20)
a=coherent(20,6)
c_a=cre*a
cc_a=cre*c_a
wigner_fock_distribution(c_a)
wigner_fock_distribution(cc_a)
```

```python
ani=destroy(20)
a=ani*coherent(20,6)
b=ani*a
wigner_fock_distribution(a)
wigner_fock_distribution(b)
```
